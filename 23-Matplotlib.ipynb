```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
ageList = [20,25,30,35,40,55,60]
```

```python
weightList = [70,75,90,85,95,93,100]
```

```python
ageListNumpy = np.array(ageList)
weightListNumpy = np.array(weightList)
```

```python
ageListNumpy
```

```python
weightListNumpy
```

```python
plt.plot(ageListNumpy, weightListNumpy, "r")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age - Weight Graph")
```

```python
myNumpy1 = np.linspace(0,15,20)
```

```python
myNumpy1
```

```python
myNumpy2 = myNumpy1 ** 3
```

```python
myNumpy2
```

```python
plt.plot(myNumpy1, myNumpy2, "g-*")
```

```python
plt.subplot(1,2,1) #1 row, 2 column, 1.graph
plt.plot(myNumpy1, myNumpy2, "r--")
plt.subplot(1,2,2)
plt.plot(myNumpy2, myNumpy1, "g*")
```

```python
myFigure = plt.figure()
myAxes = myFigure.add_axes([1,1,0.6,0.6])
myAxes.plot(myNumpy1, myNumpy2, "g")
myAxes.set_xlabel("X Data")
myAxes.set_ylabel("Y Data")
myAxes.set_title("Graph Title")
```

```python
myFigure2 = plt.figure()

myAxes3 = myFigure2.add_axes([0.1,0.1,0.9,0.9])
myAxes3.plot(myNumpy1, myNumpy2, "r*-")
myAxes3.set_xlabel("X Data Large")
myAxes3.set_ylabel("Y Data Large")
myAxes3.set_title("Large Graph")

myAxes2 = myFigure2.add_axes([0.3,0.5,0.4,0.4])
myAxes2.plot(myNumpy1, myNumpy2, "g")
myAxes2.set_xlabel("X Data Small")
myAxes2.set_ylabel("Y Data Small")
myAxes2.set_title("Small Graph")
```

```python
(myFig, myAxe) = plt.subplots()
myNumpy3 = np.linspace(0, 15, 20)
myNumpy4 = myNumpy3 ** 2
myAxe.plot(myNumpy3, myNumpy4, color="#CEA623", alpha= 0.5)
myAxe.plot(myNumpy4, myNumpy3, color="#35CE23", alpha= 0.3)
```

```python
(newFig, newAxe) = plt.subplots()
newAxe.plot(myNumpy3, myNumpy3 + 2, color="#8323CE", linewidth = 2.0)
newAxe.plot(myNumpy3, myNumpy3 + 4, color="#23CE38", linewidth = 1.5)
newAxe.plot(myNumpy3, myNumpy3 + 6, color="blue", linestyle=":")
newAxe.plot(myNumpy3, myNumpy3 + 7, color="red", linestyle=":", marker="o")
newAxe.plot(myNumpy3, myNumpy3 + 7, color="black", linestyle=":", marker="+", markersize=8)
```
